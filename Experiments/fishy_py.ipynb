```python
# Recreate Visuals
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%config InlineBackend.figure_format = 'retina'
```

```python
pickerel = pd.read_csv('https://raw.githubusercontent.com/palmorezm/misc/master/Experiments/Pickerel.csv')
pickerel.head()
```

```python
fish = pd.read_csv('https://raw.githubusercontent.com/palmorezm/misc/master/Experiments/Pickerel.csv')[['SampleID','Length','Weight','O2']]
fish.head()
```

```python
sns.scatterplot(x = 'Length',
                y = 'Weight', 
                data = fish)
```

```python
sns.scatterplot(x = 'Length',
                y = 'Weight', 
                data = fish, 
                hue = 'O2')
```

```python
plt.plot('Weight', data = fish)
plt.plot('O2', data = fish)
plt.plot('Length', data = fish)
```
